```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return 10 * np.cos(x) + np.exp(x)
```

```python
def dichotomy(a, b, e, ee):
    while True:
        x = (a + b)/2
        y1 = function(x-ee)
        y2 = function(x+ee)
        if y1 <= y2:
            b = x + ee
        else:
            a = x - ee
        
        if b - a <= 2 * e:
            return (a + b)/2
```

```python
a = -5
b = 10
e = 0.0001
ee = 0.000001
```

```python
min_x = dichotomy(a, b, e, ee)
x = np.arange(a, b+1)
y = []
[y.append(function(i)) for i in range(a, b+1)]
print(y)
```

```python
plt.plot(x, y, color = 'purple')
plt.scatter(min_x, function(min_x), color = 'red')
```
