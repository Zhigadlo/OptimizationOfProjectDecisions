```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return (x-4)*(x-4)
```

```python
def golden_ratio_method(a,b,e,l):
    x1 = b - (b-a)/l
    x2 = a + (b-a)/l
    y1 = function(x1)
    y2 = function(x2)
    while True:
        if y1 <= y2:
            b = x2
            x2 = x1
            y2 = y1
            x1 = a + b - x2
            y1 = function(x1)
        else:
            a = x1 
            x1 = x2
            y1 = y2
            x2 = a + b - x1
            y2 = function(x1)
        
        if b - a <= e:
            if y1 < y2:
                return x1
            else:
                return x2

```

```python
a = -3
b = 14
e = 0.0001
l = 1.6180339887
```

```python
min_x = golden_ratio_method(a, b, e, l)
x = np.arange(a, b+1)
y = []
[y.append(function(i)) for i in range(a, b+1)]
print(y)
```

```python
plt.plot(x, y, color = 'black')
plt.scatter(min_x, function(min_x), color = 'red')
print(min_x)
```
