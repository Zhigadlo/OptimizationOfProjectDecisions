```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return (x-4)*(x-4)
```

```python
def get_x_array(a, b, n):
    k = np.ceil(n/2)
    xk = (a + b)/2
    x = []
    for i in range(1, n):
        if i == k-1:
            x.append(xk)
        else:
            new_x = a + i*(b-a)/(n + 1)
            x.append(new_x)
    return x
```

```python
def active_block_search(a, b, e, n):
    while True:
        x = get_x_array(a, b, n)
        xf = []

        for i in x:
            xf.append(function(i))

        j = np.argmin(xf)
        a = x[j-1]
        b = x[j+1]
        xk = x[j]
        if b - a <= 2*e:
            return xk
```

```python
a = -3
b = 14
e = 0.001
n = 20
```

```python
x = get_x_array(a, b, n)
y = []
[y.append(function(i)) for i in x]
min_x = active_block_search(a, b, e, n)
plt.plot(x, y, color='blue')
plt.scatter(min_x, function(min_x), color='red')
print(min_x)
```
