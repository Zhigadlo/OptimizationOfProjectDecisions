```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return (x-4)*(x-4)
```

```python
def der_func(x):
    return 2*x - 8
```

```python
def tangent_method(a, b, e):
    aa = a
    bb = b
    while True:
        
        if bb - aa <= 2 * e:
            return (aa + bb)/2
        else:
            y1 = function(aa)
            y2 = function(bb)
            z1 = der_func(aa)
            z2 = der_func(bb)

            c = ((bb*z2-aa*z1)-(y2-y1))/(z2-z1)
            z = der_func(c)
            if z == 0:
                return c
            elif z < 0:
                aa = c
            else:
                bb = c
```

```python
a = -3
b = 15
e = 0.01
```

```python
min_x = tangent_method(a, b, e)
min_y = function(min_x)
x = np.arange(a, b+1)
y = []
[y.append(function(i)) for i in range(a, b+1)]
print(y)
```

```python
plt.plot(x, y, color='gray')
plt.scatter(min_x, min_y, color='red')
print(min_x)
```
