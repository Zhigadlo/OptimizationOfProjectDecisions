```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return 10 * np.cos(x) + np.exp(x)
```

```python
def der_func(x):
    return -10*np.sin(x) + np.exp(x)
```

```python
def tangent_method(a, b, e):
    y1 = function(a)
    y2 = function(b)
    z1 = der_func(a)
    z2 = der_func(b)

    while True:
        if b - a <= 2 * e:
            return (a + b)/2
        else:
            c = ((b*z2-a*z1)-(y2-y1))/(z2-z1)
            y = function(c)
            z = der_func(c)
            if z == 0:
                return c
            elif z < 0:
                a = c
                y1 = y
                z1 = z
            else:
                b = c
                y2 = y
                z2 - z
```

```python
a = 2
b = 15
e = 0.0001
```

```python
min_x = tangent_method(a, b, e)
min_y = function(min_x)
x = np.arange(a, b+1)
y = []
[y.append(function(i)) for i in range(a, b+1)]
print(y)
```

```python
plt.plot(x, y, color='gray')
plt.scatter(min_x, min_y, color='red')
```
